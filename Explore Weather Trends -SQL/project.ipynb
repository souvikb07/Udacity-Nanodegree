# Filtering the Data for missing values

```python
# import pandas
import pandas as pd

#create the dataframe
df = pd.read_csv('results.csv')
df.head()
```

```python
# Mean value of world temperature
mean_world_value = df['avg_temp'].mean()
print(mean_world_value)

# Mean value of City temperature
mean_city_value = df['city_avg_temp'].mean()
print(mean_city_value)

#filling the NaN value by mean
df['avg_temp'] = df['avg_temp'].fillna(mean_world_value)

df['city_avg_temp'] = df['city_avg_temp'].fillna(mean_city_value)
df.head()

# Save the filtered data
df.to_csv('final.csv')
```

# The Plot of the data

```python
# Read and create the dataframe
df = pd.read_csv('final.csv')
df.head()
```

```python
#Plot for the Comparison of Delhi Temperature and World Temperature

# Import matplotlib
import matplotlib.pyplot as plt

#Dataframe of Year and Temperature
year = df['year']
avg_temp = df['avg_temp']
city_avg = df['city_avg_temp']

# Ploting the line graph
plt.plot(year, avg_temp, color='red', label='World Temperature')
plt.plot(year, city_avg, color='blue', label='Delhi Temperature')

# Add a legend at the lower center
plt.legend(loc='center')

#Title for the Graph
plt.title('Temperature in Delhi vs World')

#labels for the axis
plt.xlabel('Year')
plt.ylabel('Temperatue')
plt.show()
```

```python
#Plot for the Comparison of Delhi Temperature and World Temperature of 10 years Average

#Create new DataFrame for the 10 Years Average Data
df1 = df[['year','MA_delhi_10years', 'MA_world_10years']]
df_10years=df1.dropna()

#Dataframe of Year and Temperature
year = df_10years['year']
delhi = df_10years['MA_delhi_10years']
world = df_10years['MA_world_10years']

# Ploting the line graph
plt.plot(year, delhi, color='red', label='World Temperature')
plt.plot(year, world, color='blue', label='Delhi Temperature')

# Add a legend at the lower center
plt.legend(loc='center')

#Title for the Graph
plt.title('Temperature in Delhi vs World(10 Years moving average)')

#labels for the axis
plt.xlabel('Year')
plt.ylabel('Temperatue')
plt.show()
```
